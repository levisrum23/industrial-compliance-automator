```python
"""
Project: Industrial Emissions Compliance Automator
Author: Anderson Rumuy
Date: January 2026
Description: This script simulates daily SO2 emission data from a pulp mill, 
             identifies regulatory violations based on a fixed threshold (120 mg/Nm3), 
             and auto-generates a priority CSV report for the Environmental Manager.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Create Dummy Data (Simulating a Pulp Mill's daily emissions)
data = {
    'Date': pd.date_range(start='2025-01-01', periods=30),
    'SO2_Emission': np.random.randint(50, 150, size=30), # Random values between 50 and 150
    'Legal_Limit': [120] * 30 # The limit is always 120
}
```

```python
df = pd.DataFrame(data)
```

```python
# 2. visualising the Data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['SO2_Emission'], label='Actual SO2 Emission', color='blue', marker='o')
plt.plot(df['Date'], df['Legal_Limit'], label='Legal Limit (120)', color='red', linestyle='--')
# 3. Add Context (Make it look professional)
plt.title('January 2025: SO2 Emission Compliance Check', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Emission Level (mg/Nm3)')
plt.legend()
plt.grid(True)
# 4. Show the plot
plt.show()
```

```python
# 5. Identify Violations
violations = df[df['SO2_Emission'] > df['Legal_Limit']]
print(f"WARNING: There were {len(violations)} legal violations this month.")
print(violations)
```

```python
# ... (Your previous code helps set this up)
```

```python
# 1. Add a Logic Column
# This creates a new column called 'Status' based on the limit
df['Status'] = df.apply(lambda x: 'VIOLATION' if x['SO2_Emission'] > x['Legal_Limit'] else 'Safe', axis=1) # PELAJARI lambda
```

```python
# 2. Filter ONLY the bad days
violation_report = df[df['Status'] == 'VIOLATION']
```

```python
# 3. export to CSV (The Deliverable)
# This creates a real file in your folder
violation_report.to_csv('High_Priority_Alerts.csv', index=False)
```

```python
print("--- PROCESS COMPLETE ---")
print(f"Analysis finished. Found {len(violation_report)} violations.")
print("I have generated a file named 'High_Priority_Alerts.csv' in your folder.")
print("Please open it to see the dates requiring investigation.")

# 4. Preview the data here for you
print("\nPreview of the Report:")
print(violation_report[['Date', 'SO2_Emission', 'Status']])
```
